```python
import seaborn as sns
import pandas as pd
```

```python
result_path = "results_native.csv"
results_df = pd.read_csv(result_path)
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
bfi_df = results_df[(results_df.Method == "BruteForceIdiomatic")]
for curr_n in pd.unique(bfi_df['n']):
    plt.title(f"n = {curr_n}")
    curr_df = bfi_df[bfi_df.n == curr_n]
    ax = sns.scatterplot(x="k", y="elapsed", hue="Method", data=curr_df)
    ax.legend(bbox_to_anchor=(1.5, 1.05))
    plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
heap_df = results_df[(results_df.Method == "Heap")]
for curr_n in pd.unique(heap_df['n']):
    plt.title(f"n = {curr_n}")
    curr_df = heap_df[heap_df.n == curr_n]
    ax = sns.scatterplot(x="k", y="elapsed", hue="Method", data=curr_df)
    ax.legend(bbox_to_anchor=(1.5, 1.05))
    plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
bst_df = results_df[(results_df.Method == "BST")]
for curr_n in pd.unique(bst_df['n']):
    plt.title(f"n = {curr_n}")
    curr_df = bst_df[bst_df.n == curr_n]
    ax = sns.scatterplot(x="k", y="elapsed", hue="Method", data=curr_df)
    ax.legend(bbox_to_anchor=(1.5, 1.05))
    plt.show()
```
